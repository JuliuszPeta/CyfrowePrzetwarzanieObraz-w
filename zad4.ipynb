```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal #Ten moduł jest użyteczny dla celów splotu
from PIL import Image
```

```python
#Zadanie 1
im = np.array(Image.open('zad4_1.jpg'))/255.0
plt.figure(figsize = (6, 6))
plt.imshow(im, cmap='gray')
```

```python
mask = np.ones((5,5))
mask
im_zad1 = scipy.signal.convolve2d(im, mask, 'same')
plt.figure(figsize = (12, 6))
plt.imshow(im_zad1, cmap='gray')
```

```python
#Zadanie 2
im1 = np.array(Image.open('zad4_2.jpg'))/255.0
plt.figure(figsize = (12, 6))
plt.imshow(im1, cmap='gray')
```

```python
mask1 = np.array([[ 0, -1,  0], [-1,  4, -1], [ 0, -1,  0]])
im_f = scipy.signal.convolve2d(im1, mask1, 'same')
im_zad2 = im1 + 50*im_f

plt.figure(figsize = (12, 6))
plt.imshow(im_zad2, cmap='gray')


                
```
